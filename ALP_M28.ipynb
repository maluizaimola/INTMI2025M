1 - Crie um algoritmo que receba 2 números, em variáveis diferentes (num1 - num2), depois faça as 4 operações com esses 2 números e mostre o resultado na tela.

```python
num1 = int(input("Digite um número: "))
num2 = int(input("Digite outro número: "))
```

```python
print("A soma é: ", num1+num2)
print("A subtração é: ", num1-num2)
print("A multiplicação é: ", num1*num2)
print("A divisão é: ", num1/num2)
```

2 - Crie um algoritmo que receba o nome do usuário na variável (nome), e a idade na variável (idade), e depois mostre na tela a seguinte frase:

Olá, meu nome é (nome) e eu tenho (idade) anos!

```python
#Entradas
nome = str(input("Digite seu nome: "))
idade = int(input("Digite sua idade: "))

#Saídas
print(f"Olá, meu nome é {nome} e eu tenho {idade} anos!")
```

```python
#Entradas
peso = float(input("Digite o seu peso(kg): "))
altura = float(input("Digite a sua altura(m)"))

#Processamento
imc = peso/(altura**2)

#Saídas
print(f"O seu IMC é: {imc:.2f}")
```
